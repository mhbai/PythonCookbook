```python
# 讀進json 檔

import json

T=json.load(open("data/華測8000詞表.json"))
print json.dumps(T,ensure_ascii=False)
```

```python
# 寫進 json 檔

import json
import codecs
# json.dump 可以使用多種輸出的格式
# 1. 最基本的方式
json.dump(T,open("data/test_out1.json","w"))

# 2. 如果存中文資料，可以設定 ensure_ascii=False, 但 open 檔案必需要用 codecs 提供的開檔
#    才能夠指定中文的編碼方式
json.dump(T,codecs.open("data/test_out2.json","w",encoding="utf8"),ensure_ascii=False)

# 3. 如果希望檔案排得整齊些，可以加上 indent 參數
json.dump(T,codecs.open("data/test_out3.json","w",encoding="utf8"),ensure_ascii=False,indent=4)

# 注意： 因為 open 是在參數中呼叫，所以不必使用 close，如果是先 open 再 dump ，記得要 close
#        忘記加 close 參數，可能造成檔案沒有完全輸出的問題
f=open("data/test_out4.json","w")
json.dump(T,f)
f.close()
```
